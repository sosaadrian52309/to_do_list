```python
to_do = "to_do_list.txt"
done = False
while done == False:
    choice = input("Would you like to add a task, remove a task, or view the current list? ")
    choice = choice.lower()
    if choice == "add":
        task = input("What task would you like to add? ")
        with open(to_do, "a") as file:
            file.write("- " + task + "\n")
        print("Task added.")
    elif choice == "remove":
        task = input("What task would you like to remove? ")
        with open(to_do, "r") as file:
            lines = file.readlines()
        with open(to_do, "w") as file:
            for line in lines:
                if line.strip() != "- " + task:
                    file.write(line)
                else:
                    removed = True
        if removed == True:
            print("Task removed.")
        else:
            print("Task not found.")
    elif choice == "view":
        try:
            with open(to_do, "r") as file:
                print("To Do List:\n" + file.read())
        except FileNotFoundError:
            print("No tasks in the list.")
    else:
        print("Invalid choice.")
    finish = input("Is there anything else you would like to do?)")
    finish = finish.lower()
    if finish == "no":
        done = True
file = open(to_do, "r")
print("To Do List:\n" + file.read())
file.close()
```
